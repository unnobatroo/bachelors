```python
def func_a():
    return 1
    print("Hi!")


print(func_a())  # 1
```

```python
# As soon as return 1 executes → function exits immediately
# print("Hi!") is never executed
```

```python
def func_a():
    try:
        return 1
    except:
        return 2
    finally:
        return 3


print(func_a())  #
```

```python
# try block runs → return 1 is encountered
# BUT before returning → Python executes finally
# finally contains return 3 → this overrides the previous return
```

```python
def func_a():
    try:
        return 1
    except:
        return 2
    finally:
        print("Done")


print(func_a())
```

```python
def func(event, log=[]):
    print(id(log))
    log.append(event)
    print(id(log))
    return log


print(func("User logged."))  #
print(func("User logout."))  #
```

```python
def func(event, log=" "):
    print(id(log))
    log = log + event
    print(id(log))
    return log


print(func("User logged."))  #
print(func("User logout."))  #
```

```python
def func(event, log=[]):
    log.append(event)
    return log


print(func("User logged.", []))
print(func("User logout.", []))
```

Default arguments are initialized once when the function is defined

```python
def func(event, log):
    log.append(event)
    return log


log = []
print(func("User logged.", log))
print(func("User logout.", log))
```

```python
def func(event):
    log = []
    log.append(event)
    return log


print(func("User logged."))
print(func("User logout."))
```
