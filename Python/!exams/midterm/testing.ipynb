Techniques for Debugging Logical Errors

3.Explain the logic (Rubber Duck Debugging)
4.Use print statements to track variables

6.Test with simple inputs
7.Take a break and revisit

```python
def calculate_grade(score):
    if score >= 80:
        grade = "B"
    elif score >= 90:
        grade = "A"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    return grade


scores = [85, 95, 78, 60, 72]
for score in scores:
    grade = calculate_grade(score)
    print(f"Score: {score}, Grade: {grade}")
```
