Write a function that determines whether the number is a leap year from the standard input. A leap year is any year divisible by four, except those divisible by a hundred. However, those divisible by four hundred will also be leap years. (DO NOT use if in the solution!)

```python
def is_leap_year() -> None:
    year = int(input("Enter a year to check: "))
    if year % 4 == 0 and year % 100 != 0:
        print(f"Yeay! {year} is a leap year.")
    elif year % 400 == 0:
        print(f"Yeay! {year} is a leap year.")
    else:
        print(f"Oh no! {year} is not a leap year.")

is_leap_year()
```

Write a function that calculates the sum of all prime numbers in given range.

```python
def print_primes_sum(start: int, end: int) -> None:
    def is_prime(num: int) -> bool:
        if num < 2:
            return False
        divs = [i for i in range(2, num) if num % i == 0]
        return len(divs) == 0

    print(f"The sum of prime numbers between {start} and {end} is {sum(i for i in range(start, end) if is_prime(i))}.")

print_primes_sum(2, 10)
```

Write a function that counts the total number of keys in a nested dictionary.

```python
def print_nested_dicts_count(dictionary: dict) -> None:
    def count_keys(d: dict) -> int:
        count = len(d)
        for value in d.values():
            if isinstance(value, dict):
                count += count_keys(value)
        return count

    print(f"Keys count: {count_keys(dictionary)}")

sample_dict = {'a': 1, 'b': {'c': 2, 'd': {'e': 3}}}
print(f"Sample dictionary: {sample_dict}")
print_nested_dicts_count(sample_dict)
```
