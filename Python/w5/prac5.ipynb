## Task 1: Leap Year Without Nested Branch Complexity

Goal: determine leap-year status from user input.

Reasoning:
- A year is leap if divisible by 400, or divisible by 4 but not by 100.
- The condition can be expressed as a boolean formula:
  `(year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)`
- Separating input parsing from the condition keeps the function readable.

```python
def is_leap_year() -> None:
    year = int(input("Enter a year to check: "))
    if year % 4 == 0 and year % 100 != 0:
        print(f"Yeay! {year} is a leap year.")
    elif year % 400 == 0:
        print(f"Yoohoo! {year} is a leap year.")
    else:
        print(f"Oh no! {year} is not a leap year.")

is_leap_year()
```

## Task 2: Sum of Primes in a Range

Reasoning:
- Build a helper `is_prime` to isolate primality logic.
- Filter numbers in the range using that helper.
- Sum only those values that pass the prime test.

This keeps complexity in one place and makes testing easier.

```python
def print_primes_sum(start: int, end: int) -> None:
    def is_prime(num: int) -> bool:
        if num < 2:
            return False
        divs = [i for i in range(2, num) if num % i == 0]
        return len(divs) == 0

    print(
        f"The sum of prime numbers between {start} and {end} is {sum(i for i in range(start, end) if is_prime(i))}."
    )

print_primes_sum(2, 10)
```

## Task 3: Count Keys in Nested Dictionaries

Reasoning:
- Use recursion because nested dictionaries have self-similar structure.
- At each level, count current keys, then recurse into dictionary values.
- Base behavior naturally occurs when a value is not a dictionary.

This guarantees every key in every nested layer is counted exactly once.

```python
def print_nested_dicts_count(dictionary: dict) -> None:
    def count_keys(d: dict) -> int:
        count = len(d)

        for value in d.values():
            if isinstance(value, dict):
                count += count_keys(value)

        return count

    print(f"Keys count: {count_keys(dictionary)}")

sample_dict = {"a": 1, "b": {"c": 2, "d": {"e": 3}}}
print(f"Sample dictionary: {sample_dict}")
print_nested_dicts_count(sample_dict)
```

## Task 5: Reverse Number Recursively (Arithmetic Only)

Reasoning:
- Recursive state carries two values: remaining digits and reversed prefix.
- Step: move last digit from `num` into `rev_num`.
- Base case: when `num == 0`, accumulated reverse is complete.

Using `% 10` and `// 10` satisfies the arithmetic-only constraint.

```python
def reverse(num: int, rev_num=0) -> int:
    if num == 0:
        return rev_num

    rev_num = rev_num * 10 + num % 10
    return reverse(num // 10, rev_num)

number = 123
print(f"The reverse of {number} is {reverse(123)}")
```

## Task 6: Recursive Palindrome Check (Letters Only)

Reasoning:
- Normalize input first: keep only letters, lowercase them.
- Recursive check compares first and last characters.
- If they match, recurse on the substring without both ends.
- Base case: length 0 or 1 is automatically palindrome.

Normalization ensures punctuation/spaces do not affect correctness.

```python
def is_palindrome(txt: str) -> bool:
    cleared_txt = "".join(letter for letter in txt if letter.isalpha()).lower()

    def check_recursive(string: str) -> bool:
        if len(string) <= 1:
            return True
        if string[0] == string[-1]:
            return check_recursive(string[1:-1])
        return False

    return check_recursive(cleared_txt)

word = "many men wish death upon me"
if is_palindrome(word):
    print(f'Yeppie! The word "{word}" is a palindrome.')
else:
    print(f'Oops! The word "{word}" is not a palindrome.')
```
