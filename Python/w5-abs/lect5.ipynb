# Part 2: Basic Error Handling Examples



## Handle Invalid Input

Write code that asks the user for a number and prints it.
If the user enters text, display `"Invalid number"`.


```python
try:
    number = int(input("Enter a number: "))
    print("You entered:", number)
except ValueError:
    print("That was not a valid Integer.")
```


## Handle Division Errors
Ask the user for a number and divide 10 by that number.
Handle both invalid input and division by zero.


```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
    print("Result:", result)
except ZeroDivisionError:
    print("You cannot divide by zero.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```


## Using else
Runs only if no exception occurs.


```python
try:
    x = int(input("Enter an integer: "))
except ValueError:
    print("Invalid input")
else:
    print("Square:", x * x)
```


## Using finally
Always runs regardless of success or failure.


# Part 5: Saving Data Using JSON


```python
import json

student = {"name": "Alice", "id": 101, "courses": ["Math", "Physics"]}

with open("student.json", "w") as f:
    json.dump(student, f, indent=4)
```

```python
with open("student.json", "r") as f:
    data = json.load(f)

print(data)
```


# Part 6: List Comprehensions


```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)
```


# Dictionary Comprehension


```python
square_dict = {x: x**2 for x in range(5)}
print(square_dict)
```


# Part 7: Mini Project: Simple To‑Do App


```python
tasks = []

while True:
    command = input("Enter command (add/view/exit): ")

    if command == "add":
        task = input("Task description: ")
        tasks.append(task)

    elif command == "view":
        for t in tasks:
            print("-", t)

    elif command == "exit":
        with open("tasks.txt", "w") as f:
            for task in tasks:
                f.write(task + "\n")
        break
```
