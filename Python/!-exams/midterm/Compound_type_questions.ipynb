```python
grades = {'anna':'A', 'David':'B', 'Betta':'A'}
lst =[]
for grade in grades:
    lst += grades[grade]
    
print(lst) # 
```

```python
lst =[]
for grade in grades.keys():
    lst += grades[grade]
    
print(lst)  # 
```

```python
for grade in grades:
    print(grade)
```

```python
for grade in grades.keys():
    print(grade)
```

```python
lst =[]
for val in grades.values():
    lst += val
print(lst) # 
```
