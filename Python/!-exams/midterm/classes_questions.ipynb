```python
class A:
    counter = 0 # class variable; shared all instances of this class
    def __init__(self, a):
        self.a = a    # instance
        A.counter += 1


ob1 = A(10)  
ob2 = A(20)  


print(ob1.a)
print(ob2.a)

print(A.counter) # 2
print(ob1.counter) # 2
print(ob2.counter) # 2 

  
```

```python
class Example:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def swap_values(self):
        print("Before swapping:", self.x, self.y)

        temp = self.x      # store original value of x
        self.x = self.y    # assign y to x
        self.y = temp      # restore x into y

        print("After swapping:", self.x, self.y)


obj = Example(3, 7)
obj.swap_values()
```

```python
class Example:
    def __init__(self):
        pass
    
    def show_square(self, number):
        number = 5
        result = number * number  
        print("The square is:", result)

obj = Example()
obj.show_square(5) 
```

```python
# result is a local variable — it exists only inside the show_square() method.
# It is created when the method starts running.
# It is destroyed when the method finishes.
```

```python
class Counter:
    count = 0
    def __init__(self, count):
        self.count = count
        Counter.count += 1
        
    def inc(self):
        self.count += 1
        

a = Counter(2) # a.count = 2
b = Counter(1) #  b.count =1
a.inc() #  a.count =3
a.inc() # a.count = 4
print(a.count, b.count, Counter.count)  #  4, 1, 2
```

How (and where) you initialize a list in Python directly determines 
whether it will be shared between instances or unique to each one.

```python
class Coordinate:
    def __init__(self,x=1,y=2):
        self.x = x
        self.y = y
        
    def setter(self,x, y):
        self.x = x
        self.y = y
        
a = Coordinate()
c = Coordinate()

print(a.x, c.x) # 

a.setter(4,5)
print(a.x, c.x) #  
```

```python
Python provides special (dunder) methods that define object behavior.
You can override them in your own classes to customize how your objects behave with built-in operations.
```

```python
class Point:
    def __init__(self, x):
        self.x = x
        
    def __eq__(self, other): 
        return self.x == other.x

p1 = Point(1)
p2 = Point(1)

print(p1 == p2)
```
