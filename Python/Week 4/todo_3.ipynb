### Iteration 6

**Feature 6.1 (tech debts):** In Iteration 5, we built a simple To-Do application that stored tasks in a list and allowed commands like “add” and “view” in a loop. Now, we will refactor that app to improve its structure using what we learned about dictionaries and functions.

### Why Refactor?

 As programs grow, organizing code into functions makes it more maintainable. Using a dictionary can make data management more flexible (for example, assigning IDs to tasks or storing additional info per task). Refactoring will help us avoid repetitive code and make future enhancements easier.

#### New Approach:

We will keep a dictionary of tasks instead of a list. Each task will have an ID (key) and a description (value). This allows us to reference tasks by an identifier.

We will create functions for key operations: e.g., add_task to add a new task to the dictionary, and view_tasks to display all tasks. The main loop will call these functions based on user commands, rather than containing all logic inline.

Improved Structure: Breaking the functionality into functions (add_task, view_tasks, etc.) means each part of the code has a single responsibility. The dictionary of tasks provides quick lookup by task ID and can be extended to hold more info (like status) later if needed.

Let's start refactoring our to-do app!
The easiest way is to copy-paste the code from Iteration 5 and then modify it step by step. The first step is to change the data structure from a list to a dictionary. Each task will have an internal key (int) and the value is the description of the task.

The first version of the functions will be created by copy-pasting the relevant code parts from the main loop.

The code after refactoring should look like this:

```python
# Initialize an empty dictionary for tasks 
tasks = {}

def add_task(description):
    """Add a new task to the tasks dictionary with a unique ID."""
    task_id = len(tasks) + 1            # generate a new ID (1, 2, 3, ...)
    tasks[task_id] = description        # store the task
    print(f'Task "{description}" added with ID {task_id}.')

def view_tasks():
    """Print all tasks with their internal IDs."""
    if not tasks:
        print("No tasks yet.")
    else:
        for tid in tasks:
            print(f"{tid}. {tasks[tid]}")

# Main application loop
while True:
    command = input("Enter command (add, view, exit): ")
    if command == "add":
        desc = input("Enter task description: ")
        add_task(desc)            # use function to add task
    elif command == "view":
        view_tasks()              # use function to display tasks
    elif command == "exit":
        print("Goodbye!")
        break
    else:
        print("Unknown command. Please type 'add', 'view', or 'exit'.")
```

**BUT** Let's refactor the code to improve its structure and maintainability based on the following principles:

- **Keep logic separate from I/O**

  - Logic functions: take inputs, return data (no print / input inside).

  - Main loop: handles input() and print() only.

- **Single Responsibility**

  - Each function should do one thing (e.g., “add a task”, “format a task list”).

- **Return data, not sentences**

  - Logic returns dictionaries, lists, booleans, or None.

  - A small “render” function turns data into human-readable text.

```python
tasks = {}

def add_task(description):
    """Add a new task to the tasks dictionary with a unique ID. Returns (task_id, task_data)."""
    task_id = len(tasks) + 1            # generate a new ID (1, 2, 3, ...)
    tasks[task_id] = description        # store the task
    return (task_id, description)

def get_all_tasks():
    """Return the tasks dictionary or None if empty."""
    return tasks if tasks else None

def render_task_added(task_id, description):
    """Render a message for a newly added task."""
    return f'Task "{description}" added with ID {task_id}.'

def render_task_list(tasks_dict):
    """Render all tasks as a formatted string."""
    if not tasks_dict:
        return "No tasks yet."
    else:
        result = []
        for tid in tasks_dict:
            result.append(f"{tid}. {tasks_dict[tid]}")
        return "\n".join(result)

# Main application loop
while True:
    command = input("Enter command (add, view, exit): ")
    if command == "add":
        desc = input("Enter task description: ")
        task_id, task_data = add_task(desc)    # get data from logic function
        message = render_task_added(task_id, task_data)  # render the message
        print(message)                         # main loop handles printing
    elif command == "view":
        tasks_data = get_all_tasks()           # get data from logic function
        message = render_task_list(tasks_data) # render the message
        print(message)                         # main loop handles printing
    elif command == "exit":
        print("Goodbye!")
        break
    else:
        print("Unknown command. Please type 'add', 'view', or 'exit'.")
```

**Feature 6.2:** Add status ("completed", "pending") and ID to each task when it is created.

Additional principles:

- **Centralize constants**

  - If you have fixed values (like allowed statuses), define them in one place (e.g., a global set or list) to avoid duplication and make updates easier.

- **Never trust user input**

  - Always validate and sanitize inputs from users to prevent errors or security issues.

```python
# Now that task has internal structure, it has an ID, description and status
# we need to change its data structure to a dictionary.
# example task entry: {1: {"description": "Buy groceries", "status": "pending"}}

# Centralize constants - define allowed statuses in one place
ALLOWED_STATUSES = {"pending", "completed"}  # using a set for fast lookup
DEFAULT_STATUS = "pending"

tasks = {}

def validate_status(status):
    """Validate user input for status. Returns valid status or default."""
    if not status:  # empty string or None
        return DEFAULT_STATUS
    
    status_clean = status.strip().lower()  # sanitize input
    if status_clean in ALLOWED_STATUSES:
        return status_clean
    else:
        return DEFAULT_STATUS  # fallback to safe default

def validate_description(description):
    """Validate and sanitize task description."""
    if not description:  # empty string or None
        return "Untitled Task"  # safe default
    
    # Basic sanitization - remove extra whitespace
    desc_clean = description.strip()
    if len(desc_clean) == 0:
        return "Untitled Task"
    
    return desc_clean

def add_task(description, status=DEFAULT_STATUS):
    """Add a new task to the tasks dictionary with a unique ID and status. Returns (task_id, task_data)."""
    # Validate and sanitize inputs
    clean_description = validate_description(description)
    clean_status = validate_status(status)
    
    task_id = len(tasks) + 1            # generate a new ID (1, 2, 3, ...)
    task_data = {"description": clean_description, "status": clean_status}
    tasks[task_id] = task_data          # store the task
    return (task_id, task_data)

def get_all_tasks():
    """Return the tasks dictionary or None if empty."""
    return tasks if tasks else None

def render_task_added(task_id, task_data):
    """Render a message for a newly added task."""
    return f'Task "{task_data["description"]}" added with ID {task_id} with status "{task_data["status"]}".'

def render_task_list(tasks_dict):
    """Render all tasks as a formatted string."""
    if not tasks_dict:
        return "No tasks yet."
    else:
        result = []
        for tid in tasks_dict:
            task = tasks_dict[tid]
            result.append(f"{tid}. {task['description']}, Status: {task['status']}")
        return "\n".join(result)

def render_allowed_statuses():
    """Render the list of allowed statuses for user guidance."""
    return f"Allowed statuses: {', '.join(sorted(ALLOWED_STATUSES))} (default: {DEFAULT_STATUS})"

# Main application loop
while True:
    command = input("Enter command (add, view, exit): ")
    if command == "add":
        desc = input("Enter task description: ")
        print(render_allowed_statuses())  # guide user on valid inputs
        status = input("Enter task status (or press Enter for default): ")
        
        task_id, task_data = add_task(desc, status)    # get data from logic function
        message = render_task_added(task_id, task_data)  # render the message
        print(message)                         # main loop handles printing
    elif command == "view":
        tasks_data = get_all_tasks()           # get data from logic function
        message = render_task_list(tasks_data) # render the message
        print(message)                         # main loop handles printing
    elif command == "exit":
        print("Goodbye!")
        break
    else:
        print("Unknown command. Please type 'add', 'view', or 'exit'.")
```

**Feature 7.1:** Add a ”find” command to search for a task by ID.

```python
ALLOWED_STATUSES = {"pending", "completed"}  # using a set for fast lookup
DEFAULT_STATUS = "pending"

tasks = {}

def validate_status(status):
    """Validate user input for status. Returns valid status or default."""
    if not status:  # empty string or None
        return DEFAULT_STATUS
    
    status_clean = status.strip().lower()  # sanitize input
    if status_clean in ALLOWED_STATUSES:
        return status_clean
    else:
        return DEFAULT_STATUS  # fallback to safe default

def validate_description(description):
    """Validate and sanitize task description."""
    if not description:  # empty string or None
        return "Untitled Task"  # safe default
    
    # Basic sanitization - remove extra whitespace
    desc_clean = description.strip()
    if len(desc_clean) == 0:
        return "Untitled Task"
    
    return desc_clean

def validate_task_id(id_str):
    """Validate and convert task ID string to integer. Returns None if invalid."""
    if not id_str:
        return None
    
    id_clean = id_str.strip()
    # Simple validation - check if it's all digits
    if id_clean.isdigit():
        return int(id_clean)
    else:
        return None

def add_task(description, status=DEFAULT_STATUS):
    """Add a new task to the tasks dictionary with a unique ID and status. Returns (task_id, task_data)."""
    # Validate and sanitize inputs
    clean_description = validate_description(description)
    clean_status = validate_status(status)
    
    task_id = len(tasks) + 1            # generate a new ID (1, 2, 3, ...)
    task_data = {"description": clean_description, "status": clean_status}
    tasks[task_id] = task_data          # store the task
    return (task_id, task_data)

def get_all_tasks():
    """Return the tasks dictionary or None if empty."""
    return tasks if tasks else None

def find_task_by_id(task_id):
    """Find and return a task by its ID. Returns task data or None if not found."""
    return tasks.get(task_id)

def render_task_added(task_id, task_data):
    """Render a message for a newly added task."""
    return f'Task "{task_data["description"]}" added with ID {task_id} with status "{task_data["status"]}".'

def render_task_list(tasks_dict):
    """Render all tasks as a formatted string."""
    if not tasks_dict:
        return "No tasks yet."
    else:
        result = []
        for tid in tasks_dict:
            task = tasks_dict[tid]
            result.append(f"{tid}. {task['description']}, Status: {task['status']}")
        return "\n".join(result)

def render_task_found(task_id, task_data):
    """Render a single task's information."""
    if task_data:
        return f"Task {task_id}: {task_data['description']}, Status: {task_data['status']}"
    else:
        return f"No task found with ID {task_id}"

def render_allowed_statuses():
    """Render the list of allowed statuses for user guidance."""
    return f"Allowed statuses: {', '.join(sorted(ALLOWED_STATUSES))} (default: {DEFAULT_STATUS})"

# Main application loop
while True:
    command = input("Enter command (add, view, find <task_id>, exit): ")
    if command == "add":
        desc = input("Enter task description: ")
        print(render_allowed_statuses())  # guide user on valid inputs
        status = input("Enter task status (or press Enter for default): ")
        
        task_id, task_data = add_task(desc, status)    # get data from logic function
        message = render_task_added(task_id, task_data)  # render the message
        print(message)                         # main loop handles printing
    elif command == "view":
        tasks_data = get_all_tasks()           # get data from logic function
        message = render_task_list(tasks_data) # render the message
        print(message)                         # main loop handles printing
    elif command.startswith("find"):
        # Parse the command to extract task ID
        command_parts = command.split()
        if len(command_parts) >= 2:
            id_str = command_parts[1]
            task_id = validate_task_id(id_str)
            if task_id is not None:
                task_data = find_task_by_id(task_id)  # get data from logic function
                message = render_task_found(task_id, task_data)  # render the message
                print(message)                     # main loop handles printing
            else:
                print(f"Invalid task ID: '{id_str}'. Please enter a valid number.")
        else:
            print("Usage: find <task_id> (e.g., 'find 2')")
    elif command == "exit":
        print("Goodbye!")
        break
    else:
        print("Unknown command. Please type 'add', 'view', 'find <task_id>', or 'exit'.")
```

**Feature 7.2:** Add a ”complete” command to mark a task as completed by ID.

```python
ALLOWED_STATUSES = {"pending", "completed"}  # using a set for fast lookup
DEFAULT_STATUS = "pending"

tasks = {}

def validate_status(status):
    """Validate user input for status. Returns valid status or default."""
    if not status:  # empty string or None
        return DEFAULT_STATUS
    
    status_clean = status.strip().lower()  # sanitize input
    if status_clean in ALLOWED_STATUSES:
        return status_clean
    else:
        return DEFAULT_STATUS  # fallback to safe default

def validate_description(description):
    """Validate and sanitize task description."""
    if not description:  # empty string or None
        return "Untitled Task"  # safe default
    
    # Basic sanitization - remove extra whitespace
    desc_clean = description.strip()
    if len(desc_clean) == 0:
        return "Untitled Task"
    
    return desc_clean

def validate_task_id(id_str):
    """Validate and convert task ID string to integer. Returns None if invalid."""
    if not id_str:
        return None
    
    id_clean = id_str.strip()
    # Simple validation - check if it's all digits
    if id_clean.isdigit():
        return int(id_clean)
    else:
        return None

def add_task(description, status=DEFAULT_STATUS):
    """Add a new task to the tasks dictionary with a unique ID and status. Returns (task_id, task_data)."""
    # Validate and sanitize inputs
    clean_description = validate_description(description)
    clean_status = validate_status(status)
    
    task_id = len(tasks) + 1            # generate a new ID (1, 2, 3, ...)
    task_data = {"description": clean_description, "status": clean_status}
    tasks[task_id] = task_data          # store the task
    return (task_id, task_data)

def get_all_tasks():
    """Return the tasks dictionary or None if empty."""
    return tasks if tasks else None

def find_task_by_id(task_id):
    """Find and return a task by its ID. Returns task data or None if not found."""
    return tasks.get(task_id)

def complete_task_by_id(task_id):
    """Mark a task as completed by its ID. Returns updated task data or None if not found."""
    task_data = tasks.get(task_id)
    if task_data:
        tasks[task_id]["status"] = "completed"
        return tasks[task_id]  # return updated task data
    else:
        return None

def render_task_added(task_id, task_data):
    """Render a message for a newly added task."""
    return f'Task "{task_data["description"]}" added with ID {task_id} with status "{task_data["status"]}".'

def render_task_list(tasks_dict):
    """Render all tasks as a formatted string."""
    if not tasks_dict:
        return "No tasks yet."
    else:
        result = []
        for tid in tasks_dict:
            task = tasks_dict[tid]
            result.append(f"{tid}. {task['description']}, Status: {task['status']}")
        return "\n".join(result)

def render_task_found(task_id, task_data):
    """Render a single task's information."""
    if task_data:
        return f"Task {task_id}: {task_data['description']}, Status: {task_data['status']}"
    else:
        return f"No task found with ID {task_id}"

def render_task_completed(task_id, task_data):
    """Render a message for a completed task."""
    if task_data:
        return f"Task {task_id} marked as completed: {task_data['description']}"
    else:
        return f"No task found with ID {task_id}"

def render_allowed_statuses():
    """Render the list of allowed statuses for user guidance."""
    return f"Allowed statuses: {', '.join(sorted(ALLOWED_STATUSES))} (default: {DEFAULT_STATUS})"

# Main application loop
while True:
    command = input("Enter command (add, view, find <task_id>, complete <task_id>, exit): ")
    if command == "add":
        desc = input("Enter task description: ")
        print(render_allowed_statuses())  # guide user on valid inputs
        status = input("Enter task status (or press Enter for default): ")
        
        task_id, task_data = add_task(desc, status)    # get data from logic function
        message = render_task_added(task_id, task_data)  # render the message
        print(message)                         # main loop handles printing
    elif command == "view":
        tasks_data = get_all_tasks()           # get data from logic function
        message = render_task_list(tasks_data) # render the message
        print(message)                         # main loop handles printing
    elif command.startswith("find"):
        # Parse the command to extract task ID
        command_parts = command.split()
        if len(command_parts) >= 2:
            id_str = command_parts[1]
            task_id = validate_task_id(id_str)
            if task_id is not None:
                task_data = find_task_by_id(task_id)  # get data from logic function
                message = render_task_found(task_id, task_data)  # render the message
                print(message)                     # main loop handles printing
            else:
                print(f"Invalid task ID: '{id_str}'. Please enter a valid number.")
        else:
            print("Usage: find <task_id> (e.g., 'find 2')")
    elif command.startswith("complete"):
        # Parse the command to extract task ID
        command_parts = command.split()
        if len(command_parts) >= 2:
            id_str = command_parts[1]
            task_id = validate_task_id(id_str)
            if task_id is not None:
                updated_task_data = complete_task_by_id(task_id)  # get data from logic function
                message = render_task_completed(task_id, updated_task_data)  # render the message
                print(message)                     # main loop handles printing
            else:
                print(f"Invalid task ID: '{id_str}'. Please enter a valid number.")
        else:
            print("Usage: complete <task_id> (e.g., 'complete 2')")
    elif command == "exit":
        print("Goodbye!")
        break
    else:
        print("Unknown command. Please type 'add', 'view', 'find <task_id>', 'complete <task_id>', or 'exit'.")
```
