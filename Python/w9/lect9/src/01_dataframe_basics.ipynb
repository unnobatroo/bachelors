## 1. A `Series` is one labeled result

A `Series` is often the smallest useful building block in pandas.

Think of it as:
- one labeled dimension of values
- often one column by itself
- or the result of a summary such as `value_counts()`
- with labels that make explanation easier

```python
section("Series: one labeled dimension")

penguin_counts = pd.Series(
    [152, 68, 124],
    index=["Adelie", "Chinstrap", "Gentoo"],
    name="records",
)

print(penguin_counts)
print(f"\nAdelie count: {penguin_counts['Adelie']}")
```

## 2. A `DataFrame` is the working table

This is the core dataframe mental model from the lecture:
- rows represent records, events, or observations
- columns represent variables or features
- labels make access and explanation easier
- dtypes are part of the table's meaning, not just storage details

```python
section("DataFrame: the working table")

df_islands = pd.DataFrame(
    {
        "island": ["Torgersen", "Dream", "Biscoe"],
        "records": [152, 68, 124],
        "sampled": [True, True, True],
    },
    index=["Adelie", "Chinstrap", "Gentoo"],
)

print(df_islands)
print("\nColumn access:")
print(df_islands["island"])
print("\nLabel-based row access:")
print(df_islands.loc["Adelie"])
print("\nPosition-based row access:")
print(df_islands.iloc[0])
```

## 3. Read a real table and inspect it quickly

We use a small local dataset so the notebook works offline during class.

As you run the next cell, look for:
- the first few rows
- the inferred dtypes
- whether any values are missing
- what one row means in this dataset

```python
section("Read a real table and inspect it")

missions = pd.read_csv(
    data_path("space_missions.csv"),
    dtype_backend="pyarrow",
)

print("First rows:")
print(missions.head(5).to_string(index=False))
print("\nDtypes:")
print(missions.dtypes.to_string())
print("\nMissing values:")
print(missions.isna().sum().to_string())
```

## 4. Ask first questions with common access patterns

The next cell demonstrates patterns that show up constantly in pandas:
- `iloc` for a first positional check
- `loc` for label or mask-based access
- `sort_values()` after filtering
- `value_counts()` for a quick summary

### Mini class prompt
Before running the final lines, guess which agency appears most often in this small demo dataset.

```python
section("Ask first questions with common access patterns")

print("First row with iloc:")
print(missions.iloc[0])

moon_missions = missions.loc[
    missions["destination"] == "Moon",
    ["mission_name", "launch_year", "agency", "cost_million_usd"],
].sort_values("launch_year")

print("\nMoon missions with loc + sorting:")
print(moon_missions.to_string(index=False))

agency_counts = missions["agency"].value_counts()
print("\nAgency counts:")
print(agency_counts.to_string())

print(f"\nMost frequent agency in the demo data: {agency_counts.index[0]}")
```
