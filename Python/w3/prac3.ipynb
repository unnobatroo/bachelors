## Task 1: 8x8 Chessboard Pattern

Approach:
- Use two nested loops over row and column indices.
- The color depends on parity of `row + col`.
- Build one output string and print once.

Why it works:
- Adjacent cells differ by 1 in index sum, so parity alternates naturally.

```python
"""Draw an 8×8 empty chessboard using nested for loops.
Print [] for dark fields and two spaces for light fields.
Use divisibility by two to decide the field."""

str = ""
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            str += "[]"
        else:
            str += "  "
    str += "\n"
print(str)
```

## Task 2: Prime Detection and Printing

Approach:
- For a number `n`, test divisors only up to `sqrt(n)`.
- Reuse `is_prime` inside a loop over target range.

Why it works:
- If `n` has a non-trivial divisor, one of the pair must be `<= sqrt(n)`.

```python
import math

def is_prime(a):
    """Read a number and determine whether it is prime. Use a for
    loop and stop the loop early using break when a divisor is
    found."""
    for i in range(2, int(math.sqrt(a)) + 1):
        if a % i == 0:
            return False
    return True

for i in range(1, 100):
    if is_prime(i):
        print(f"AHA! The number -{i}- is prime.")
```
