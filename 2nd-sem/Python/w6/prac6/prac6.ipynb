Write a Python program that lists all files in a given dir using os.listdir() and determines
 their sizes using os.path.getsize(). Sort the files by size and write the filenames and their 
 sizes to file_sizes.txt. You may use a lambda function for sorting, but it is not required. Include 
 error handling for invalid dir paths, permission errors, and other unexpected exceptions.


Approach hint:
- identify failure points first
- keep `try` blocks small
- catch specific exceptions before generic ones


```python
import os

try:
    np = os.listdir()
    fs = []

    for n in np:
        s = os.path.getsize(n)
        fs.append((n, s))

    fs.sort(key=lambda s: s[1])

    with open("file_sizes.txt", "w") as f:
        for n, s in fs:
            f.write(f"{n} ---> {s} bytes\n")

except PermissionError:
    print("Permission denied while accessing the dir or files.")
except Exception as e:
    print(f"Unexpected error: {e}")
```

and selects only files with the .txt extension. For each .txt file, count the number of lines 
using list comprehension, store the filename and line count in a dictionary named file_reposrt,

 that cannot be opened, permission issues, and other unexpected exceptions.

- identify failure points first

- catch specific exceptions before generic ones


```python
import os

def cl(path):
    lc = 0
    with open(path, "r") as f:
        for lc, _ in enumerate(f, 1):
            pass
    return lc

try:
    fs = [
        (n, cl(n))
        for n in os.listdir()
        if n[-4:] == ".txt" and os.path.isfile(n)
    ]
    
    file_reposrt = dict(fs)
    
    with open("report.txt", "w") as f:
        for n, l in fs:
            f.write(f"{n} : {l} lines\n")

except PermissionError:
    print("Permission denied while accessing the dir or files.")
except Exception as e:
    print(f"Error: {e}")
```
